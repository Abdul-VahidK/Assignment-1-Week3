# 1-Display Given Data

```python
import pandas as pd
iris = pd.read_excel('iris.xls')
```

```python
iris
```

# 2-Display Columns in the Dataset

#### Here we displayed the names of columns in the given dataset

```python
iris.columns
```

# 3-Mean of each column of the dataset

#### we find the mean of each column in the dataset using .mean() function and print the mean value of each column

```python
mean_SL = iris['SL'].mean()
mean_SW = iris['SW'].mean()
mean_PL = iris['PL'].mean()
mean_PW = iris['PW'].mean()
```

```python
print('Mean value of the column SL:',mean_SL)
print('Mean value of the column SW:',mean_SW)
print('Mean value of the column PL:',mean_PL)
print('Mean value of the column PW:',mean_PW)
```

# 5-Visualisations of the Given Dataset

### *Boxplot of Each Column

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
plt.boxplot(iris['SL'])
plt.title('Distribution of SL')
plt.xlabel('SL')

plt.subplot(2,2,2)
plt.boxplot(iris['SW'])
plt.title('Distribution of SW')
plt.xlabel('SW')

plt.subplot(2,2,3)
plt.boxplot(iris['PL'])
plt.title('Distribution of PL')
plt.xlabel('PL')

plt.subplot(2,2,4)
plt.boxplot(iris['PW'])
plt.title('Distribution of PW')
plt.xlabel('PW')

plt.tight_layout()
```

### *Histogram of Each Column

```python
plt.figure(figsize=(8,8))

plt.subplot(2,2,1)
plt.hist(iris['SL'],rwidth=0.9)
plt.title('Distribution of SL')
plt.xlabel('SL')

plt.subplot(2,2,2)
plt.hist(iris['SW'],rwidth=0.9)
plt.title('Distribution of SW')
plt.xlabel('SW')

plt.subplot(2,2,3)
plt.hist(iris['PL'],rwidth=0.9)
plt.title('Distribution of PL')
plt.xlabel('PL')

plt.subplot(2,2,4)
plt.hist(iris['PW'],rwidth=0.9)
plt.title('Distribution of PW')
plt.xlabel('PW')

plt.tight_layout()
```

### *Strip Plot of Different Columns with the Classification in the Given Dataset

```python
import seaborn as sns
plt.figure(figsize=(9,9))

plt.subplot(2,2,1)
sns.stripplot(x='Classification',y='SL',data=iris)
plt.title('Strip Plot of Classification vs SL')

plt.subplot(2,2,2)
sns.stripplot(x='Classification',y='SW',data=iris)
plt.title('Strip Plot of Classification vs SW')

plt.subplot(2,2,3)
sns.stripplot(x='Classification',y='PL',data=iris)
plt.title('Strip Plot of Classification vs PL')

plt.subplot(2,2,4)
sns.stripplot(x='Classification',y='PW',data=iris)
plt.title('Strip Plot of Classification vs PW')

plt.tight_layout()
```
